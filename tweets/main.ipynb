```python
import pandas as pd
```

```python
tweets_data = pd.read_csv('tweets-dataset.csv', encoding='latin-1')
```

```python
tweets_data = tweets_data.set_axis(["target", "id", "date", "flag", "user", "text"], axis=1)
```

```python
tweets_data.drop(['flag', 'target'], inplace=True, axis=1)
```

```python
tweets_data.user.groupby(tweets_data.user).count().sort_values(ascending=False).head(10).plot(kind='bar', figsize=(10, 5), title='Top 10 users with most tweets')
```

```python
# tweets from the top 20 tweeters
data = tweets_data[tweets_data.user.isin(tweets_data.user.groupby(tweets_data.user).count().sort_values(ascending=False).head(20).index)]
```

```python
# separate the tweets by user and save them in a separate csv file
for user in data.user.unique():
    data[data.user == user].to_csv(f'./user_tweets/{user}.csv', index=False)
```
